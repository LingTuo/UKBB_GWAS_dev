```sos
anno <- read.table('~/Rare_Variants_MWE/annotation_file.txt')
```

```sos
colnames(anno) <- c("var", "gene", "annotation")
head(anno, 50)
```

```sos
library('tidyr')
```

```sos
anno <- separate(anno, 'gene', paste("gene", 1:4, sep="_"), sep=";", extra="drop")
anno <- separate(anno, 'var', paste("var", 1:4, sep="_"), sep=":", extra="drop", remove=FALSE)
```

```sos
head(anno)
```

```sos
library('tidyverse')
anno_new <- anno %>%
    select(gene_1,var,var_1,var_2, annotation)
```

```sos
head(anno_new, 100)
```

```sos
write.table(anno_new, file = "~/Rare_Variants_MWE/annotation_file_onegene.txt",quote = FALSE, sep = " ",row.names = FALSE,col.names = FALSE)
```

```sos
anno_new %>%
    group_by(gene_1, var_1)
```
