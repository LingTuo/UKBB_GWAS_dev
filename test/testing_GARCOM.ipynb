```sos
genecoord_test <- read.table(gzfile('/home/dc2325/software/refGene.txt.gz'), sep='\t', header=F)
names(genecoord_test) <-c("bin","name","CHR", "strand", "START", "END", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds", "score", "GENE", "cdsStartStat", "cdsEndStat", "exonFrames")
```

```sos
head(genecoord_test)
```

```sos
myvars <- c("CHR","START", "END", "GENE")
genecoord_test <- genecoord_test[myvars]
genecoord_test$CHR <- gsub('chr','', genecoord_test$CHR)
```

```sos
head(genecoord_test, 40)
```
