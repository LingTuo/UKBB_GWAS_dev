```sos
genecoord <- read.table(gzfile('/home/dc2325/software/refGene.txt.gz'), sep='\t', header=F)
names(genecoord) <-c("bin","name","CHR", "strand", "START", "END", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds", "score", "GENE", "cdsStartStat", "cdsEndStat", "exonFrames")
```

```sos
myvars <- c("CHR","START", "END", "GENE")
genecoord <- genecoord[myvars]
genecoord$CHR <- gsub('chr','', genecoord$CHR)
```

```sos
gene_test <- genecoord %>%
    filter(CHR==22) %>%
    arrange(START) 
head(gene_test)
dim(gene_test)
```

```sos
gene_unique_nogene <- unique(gene_test)
```

```sos
head(gene_unique_nogene)
dim(gene_unique_nogene)
```

```sos
gene_unique = gene_test[!duplicated(gene_test$GENE),]
head(gene_unique)
dim(gene_unique)
```

```sos
gene_unique <- gene_unique %>%
    select("CHR", "START", "END")
head(gene_unique)
```

```sos
chunk <- 200
n <- nrow(gene_unique)
r  <- rep(1:ceiling(n/chunk),each=chunk)[1:n]
d <- split(gene_unique,r)
```
