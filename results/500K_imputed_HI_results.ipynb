```sos
library('dplyr')
library('hudson')
library('ggrepel')
```

```sos
# Import relevant sumstats
gwas.1 <- read.table(gzfile('/home/dmc2245/UKBiobank/results/REGENIE_results/results_exome_data/090921_f3393_hearing_aid_200K/090321_UKBB_Hearing_aid_f3393_expandedwhite_6436cases_96601ctrl_PC1_2_f3393.regenie.snp_stats.gz'), sep='\t', header=T)
gwas.2 <- read.table(gzfile('/home/dmc2245/UKBiobank/results/REGENIE_results/results_exome_data/091421_f3393_hearing_aid_150K/090321_UKBB_Hearing_aid_f3393_expandedwhite_78848ind_150K_f3393.regenie.snp_stats.gz'), sep='\t', header=T)
```

```sos
annot_file <- read.csv(gzfile('~/UKBiobank/results/ukb23155_200Kexomes_annovar/091321_exome_bim_merge/ukb23155_chr1_chr22_091321.hg38.hg38_multianno.csv.gz'))
```

```sos
names(annot_file)[names(annot_file) == "Otherinfo1"] <- "SNP"
annot_file$Gene <- sub(";", , annot_file$Gene.refGene)
```

```sos
qqdat1 <- data.frame(SNP=gwas.1$SNP,CHR=gwas.1$CHR,POS=gwas.1$POS, pvalue=gwas.1$P,length.out=nrow(gwas.1), each=1)
qqdat1 <- qqdat1[which(qqdat1$pvalue<0.05), ]
```

```sos
qqdat2 <- data.frame(SNP=gwas.2$SNP,CHR=gwas.2$CHR,POS=gwas.2$POS, pvalue=gwas.2$P,length.out=nrow(gwas.2), each=1)
qqdat2 <- qqdat1[which(qqdat2$pvalue<0.05), ]
```

```sos
qqdat1_new <- merge(x = qqdat1, y = annot_file[ , c("SNP", "Gene.refGene", "avsnp150")], by = "SNP", all.x=TRUE)
```

```sos
#Create mirror plot
gmirror(top=qqdat1, bottom=qqdat2, tline=5e-08, bline=5e-08,
        toptitle="H-aid", bottomtitle="H-aid", log10=TRUE, opacity=0.5,
        annotate_snp = c(),
        highlight_snp = c(),
        highlight_p = c(0.0,0.0), highlighter="red", background='white',freey,
        file='/home/dmc2245/test.png', res=300)
```
