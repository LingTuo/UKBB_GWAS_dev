```sos
for i in {1..22}
do
tpl_file=~/working/bioworkflows/admin/csg.yml
pops_sos=~/working/bioworkflows/PoPS/pops.ipynb
pops_sbatch_regenie=~/project/guangyou/20220218_UKB_pops/scripts/magma_f2247_chr${i}_$(date +"%Y-%m-%d").sbatch
work_dir=~/project/guangyou/20220218_UKB_pops/magma
bfile=/home/gl2776/UKBiobank/data/exome_files/project_VCF/072721_run/plink/ukb23156_c${i}.merged.filtered.bed
gene_loc=/mnt/mfs/statgen/guangyou/data/pops-MWE/NCBI38.gene.loc
gene_annot=~/project/guangyou/20220218_UKB_pops/magma/annoate.genes.annot
pval=~/project/guangyou/20220218_UKB_pops/magma/090321_UKBB_Hearing_difficulty_f2247_expandedwhite_45502cases_96601ctrl_PC1_2_f2247.regenie.snp_stats.txt
snpcol="ID"
pvalcol="P"
ncol="N"
out=~/project/guangyou/20220218_UKB_pops/magma/UKBB_Hearing_difficulty_f2247_chr${i}

pops_args="""magma
    --cwd ${work_dir} \
    --bfile ${bfile}\
    --gene_annot ${gene_annot}\
    --pval ${pval}\
    --snpcol ${snpcol}\
    --pvalcol ${pvalcol}\
    --ncol ${ncol}\
    --container_pops /mnt/mfs/statgen/guangyou/containers/pops.sif\
    --out ${out}
"""


sos run ~/working/UKBB_GWAS_dev/admin/Get_Job_Script.ipynb csg \
    --template-file $tpl_file \
    --workflow-file $pops_sos \
    --to-script $pops_sbatch_regenie \
    --args "$pops_args"
done
```

```sos
# Combine the scripts
i=1
FILE=~/project/guangyou/20220218_UKB_pops/scripts/annotate_chr${i}_2022-02-22.sbatch
cat > ${FILE} <<- EOM
#!/bin/sh
#$ -l h_rt=36:00:00
#$ -l h_vmem=16G
#$ -N pops_2022-02-18
#$ -o /home/gl2776/project/guangyou/20220218_UKB_pops/scripts/pops_2022-02-18-$JOB_ID.out
#$ -e /home/gl2776/project/guangyou/20220218_UKB_pops/scripts/pops_2022-02-18-$JOB_ID.err  
#$ -j y
#$ -q csg.q
#$ -S /bin/bash
export PATH=$HOME/miniconda3/bin:$PATH
module load Singularity/3.5.3
sos run /home/gl2776/working/bioworkflows/PoPS/pops.ipynb \
    annotate\
    --cwd /home/gl2776/project/guangyou/20220218_UKB_pops/magma     --snp_loc /home/gl2776/UKBiobank/data/exome_files/project_VCF/072721_run/plink/ukb23156_c${i}.merged.filtered.bim   --gene_loc /mnt/mfs/statgen/guangyou/data/pops-MWE/NCBI38.gene.loc    --winup 0    --windown 0    --container_pops /mnt/mfs/statgen/guangyou/containers/pops.sif    --out /home/gl2776/project/guangyou/20220218_UKB_pops/magma/annoate \
    -c /home/gl2776/working/bioworkflows/admin/csg.yml \
    -q csg -s force \
    &> /home/gl2776/project/guangyou/20220218_UKB_pops/scripts/annotate_chr${i}_2022-02-22.log
EOM
```
