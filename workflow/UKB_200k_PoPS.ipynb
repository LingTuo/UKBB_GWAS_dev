### feature merge and selection

Use the features of brain tissues from https://github.com/FinucaneLab/gene_features/tree/master/features . It seems that the brain tissue is the only one relevent

```sos
import pandas as pd
import os 
# Gene expression features read in and merge
path="/home/gl2776/project/guangyou/20220218_UKB_pops/features/gene_features/features/"
folderlist = [f for f in os.listdir(path) if f.startswith("human_brain")]
subpath = folderlist[0]
files = [f for f in os.listdir(os.path.join(path,subpath)) if f.endswith(".txt.gz")]
files
```

```sos
folderlist
```

```sos
files = [f for f in os.listdir(os.path.join(path,subpath)) if f.endswith(".txt.gz")]
feature_df=pd.read_csv(os.path.join(path,subpath,files[0]),sep="\t",compression = "gzip")
feature_df=feature_df.add_prefix("%s_%s_"%(subpath,files[0].split(".")[0]))
feature_df=feature_df.rename(columns={"%s_%s_ENSG"%(subpath,files[0].split(".")[0]):"ENSG"})
for i in range(1,len(files)):
    x=pd.read_csv(os.path.join(path,subpath,files[i]),sep="\t",compression = "gzip")
    x=x.add_prefix("%s_%s_"%(subpath,files[0].split(".")[0]))
    x=x.rename(columns={"%s_%s_ENSG"%(subpath,files[0].split(".")[0]):"ENSG"})
    feature_df=feature_df.merge(x, on='ENSG', how='inner')
feature_df=feature_df.rename(columns={"ENSG":"ENSGID"})
feature_df
```

```sos
feature = feature_df[["ENSGID"]]
for subpath in folderlist[:6]:
    files = [f for f in os.listdir(os.path.join(path,subpath)) if f.endswith(".txt.gz")]
    feature_df=pd.read_csv(os.path.join(path,subpath,files[0]),sep="\t",compression = "gzip")
    feature_df=feature_df.add_prefix("%s_%s_"%(subpath,files[0].split(".")[0]))
    feature_df=feature_df.rename(columns={"%s_%s_ENSG"%(subpath,files[0].split(".")[0]):"ENSG"})
    for i in range(1,len(files)):
        x=pd.read_csv(os.path.join(path,subpath,files[i]),sep="\t",compression = "gzip")
        x=x.add_prefix("%s_%s_"%(subpath,files[0].split(".")[0]))
        x=x.rename(columns={"%s_%s_ENSG"%(subpath,files[0].split(".")[0]):"ENSG"})
        feature_df=feature_df.merge(x, on='ENSG', how='inner')
    feature_df=feature_df.rename(columns={"ENSG":"ENSGID"})
    feature=feature.merge(feature_df, on='ENSGID', how='inner')
feature
```
